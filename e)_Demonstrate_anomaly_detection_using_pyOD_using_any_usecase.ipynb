```python
import numpy as np
import pandas as pd
from pyod.models.iforest import IForest
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Generate synthetic temperature data
np.random.seed(42)
X = np.random.normal(loc=20, scale=5, size=(1000, 1))
X[::50] += np.random.normal(loc=20, scale=20, size=(20, 1))  # Inject anomalies

# Convert to DataFrame for ease of use
data = pd.DataFrame(X, columns=['Temperature'])

# Split the dataset
X_train, X_test = train_test_split(data, test_size=0.2, random_state=42)

# Create an Isolation Forest model
clf = IForest()

# Fit the model
clf.fit(X_train)

# Predict outliers
y_test_pred = clf.predict(X_test)

# Visualization
plt.figure(figsize=(12, 6))
plt.scatter(X_test.index, X_test['Temperature'], c=y_test_pred, cmap='coolwarm')
plt.title("Anomaly Detection in Temperature Data")
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.show()
```
